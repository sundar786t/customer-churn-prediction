#Data Preprocessing

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Load CSV
df = pd.read_csv("c:/users/lsund/downloads/Churn_Modelling.csv")

# 1. Handle missing values (if any)
df = df.fillna(df.median())  # numeric columns
df = df.dropna()              # any remaining missing rows

# 2. Encode categorical features
df['Gender'] = df['Gender'].map({'Male':0, 'Female':1})
df = pd.get_dummies(df, columns=['Geography'], drop_first=True)  # One-hot encoding

# 3. Separate features and target
X = df.drop(['CustomerID','Exited'], axis=1)
y = df['Exited']

# 4. Scale numerical variables
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
# Numeric columns only
num_cols = df.select_dtypes(include=np.number).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())

# Categorical columns
cat_cols = df.select_dtypes(include='object').columns
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

```
