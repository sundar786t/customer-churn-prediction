```python
import pandas as pd

# CSV load 
df = pd.read_csv("c:/users/lsund/downloads/Churn_Modelling.csv")

# Outlier detect  function (IQR method)
def detect_outliers_iqr(data):
    outliers = {}
    for col in data.select_dtypes(include=['float64','int64']).columns:
        Q1 = data[col].quantile(0.25)
        Q3 = data[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers[col] = data[(data[col] < lower_bound) | (data[col] > upper_bound)][col]
    return outliers

outliers = detect_outliers_iqr(df)

# Outliers print 
for col, values in outliers.items():
    print(f"\nColumn: {col}")
    print(values)
```

```python
from scipy import stats
import numpy as np

# Numerical columns 
num_cols = df.select_dtypes(include=['float64','int64']).columns

z_scores = np.abs(stats.zscore(df[num_cols]))
outliers_z = (z_scores > 3)

# Outliers indices print
for i, col in enumerate(num_cols):
    print(f"\nColumn: {col}")
    print(df[col][outliers_z[:,i]])
```
